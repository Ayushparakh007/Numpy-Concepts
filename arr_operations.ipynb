```python
a=np.array([1,2,3,4,5])
b=np.array([6,7,8,9,10])
```

```python
np.vstack((a,b))
```

```python
np.hstack((a,b)) #horizontal stack
```

```python
np.column_stack((a,b)) #column stack is used to stack 1D arrays as columns into a 2D array. It takes each input array and stacks them column-wise, resulting in a 2D array where each input array becomes a column.
```
